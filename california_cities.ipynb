```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head(1)


cities["area_total_km2"].sort_values(ascending=False)
```

```python
lat,lon = cities['latd'],cities['longd']
population,area = cities['population_total'],cities['area_total_km2'];
plt.style.use('seaborn');
plt.figure(figsize=(10,8));
plt.scatter(lon,lat,c=np.log10(population),cmap='viridis',s=area,linewidths=0,alpha=0.5)
plt.axis('equal')
plt.xlabel='longlatitude'
plt.ylabel='latitude';
plt.colorbar(label='population');
plt.clim(3,7)

area_range=[50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,alpha=0.4, label=(str(area)+'km^2'))


plt.legend(labelspacing=1,title='City')
plt.title('California Cities: Population and Area distribution');
```
